```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from latex_conf import build_latex_confusion_matrix
```

```python
df_phish = pd.read_csv("result_custom_.txt", sep="\t")
df_phish["is_phish"] = 1
df_benign = pd.read_csv("result_custom_benign.txt", sep="\t")
df_benign["is_phish"] = 0

df = df_phish.append(df_benign, ignore_index=True)

df_phish["phish_category"] = np.where(df_phish['true_brand'] == df_phish['pred_brand'], 1, 0)
df_rec = df_phish.append(df_benign, ignore_index=True)
```

```python
confusion_matrix = pd.crosstab(df['is_phish'], df['phish_category'], rownames=['Actual'], colnames=['Predicted'], margins=True)
confusion_matrix_rec = pd.crosstab(df_rec['is_phish'], df_rec['phish_category'], rownames=['Actual'], colnames=['Predicted'], margins=True)
```

```python
confusion_matrix
```

```python
confusion_matrix_rec
```

```python
tp, tp_rec = confusion_matrix[1][1], confusion_matrix_rec[1][1]
fp, fp_rec = confusion_matrix[1][0], confusion_matrix_rec[1][0]
fn, fn_rec = confusion_matrix[0][1], confusion_matrix_rec[0][1]
ppv = tp / (tp + fp)
ppv_rec = tp_rec / (tp_rec + fp_rec)
print(f"Precision: {ppv}, Precision brand: {ppv_rec}")
```

```python
tpr = tp / (tp + fn)
tpr_rec = tp_rec / (tp_rec + fn_rec)
print(f"Sensitivity: {tpr}, Sensitivity brand: {tpr_rec}")
```

```python
print(build_latex_confusion_matrix(confusion_matrix))
```
